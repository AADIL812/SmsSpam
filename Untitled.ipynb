```python
import numpy as np
import pandas as pd
import nltk
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv('spam.csv', encoding='latin-1')

# 1. Data Cleaning 

# Drop last 3 columns
df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace=True)
print(df.sample(5))

# Rename columns
df.rename(columns={'v1': 'target', 'v2': 'text'}, inplace=True)
print(df.sample(5))

# Encode target labels ('ham' = 0, 'spam' = 1)
encoder = LabelEncoder()
df['target'] = encoder.fit_transform(df['target'])
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Check for duplicate values
print(df.duplicated().sum())

# Drop duplicate rows
df = df.drop_duplicates(keep='first')
print(df.duplicated().sum())

# 2. EDA (Exploratory Data Analysis)
print(df['target'].value_counts())

# Plot pie chart for class distribution
plt.pie(df['target'].value_counts(), labels=['ham', 'spam'], autopct="%0.2f")
plt.show()

# Data is imbalanced as there are fewer spam samples

# Ensure necessary NLTK resources are downloaded
nltk.download('punkt_tab')

# 3. Feature Engineering
df['num_char'] = df['text'].apply(len)
df['num_word'] = df['text'].apply(lambda x: len(nltk.word_tokenize(x)))
df['num_sent'] = df['text'].apply(lambda x: len(nltk.sent_tokenize(x)))

# Show a sample of the DataFrame with new features
print(df.head())


df[['num_char','num_word','num_sent']].describe()

import seaborn as sns
sns.heatmap(df.select_dtypes(include=['number']).corr(),annot=True)

#since correlation between target and num_char have higher correlation we use num_char for making the model

```

```python
##3 Data pre processing
 # Lowercase
 # Tokenization
 # Removing special characters
 # Removing stop words and punctuators
 #stemming
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import string

# Load stop words and punctuation
stop = set(stopwords.words('english'))  # Use a set for faster lookups
punc = set(string.punctuation)  # Convert to set for efficient checks
ps=PorterStemmer();

def transform_text(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Tokenize the text
    text = nltk.word_tokenize(text)
    
    # Filter out stop words, punctuation, and non-alphanumeric tokens
    filtered_text = [word for word in text if word.isalnum() and word not in stop and word not in punc]

    text=filtered_text[:];
    filtered_text.clear()
    for i in text:
        filtered_text.append(ps.stem(i))
    
    return " ".join(filtered_text)
```

```python
transform_text('hi. Aadil8@. you are bad. Are you dancing ')
```

```python
df['transform_text']=df['text'].apply( lambda x:transform_text(x)) 
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

spam_most = []

# Iterate through the list of spam messages (target == 1)
for msg in df[df['target'] == 1]['transform_text'].tolist():
    for word in msg.split():
        spam_most.append(word)

# Find the 30 most common words
most_common = Counter(spam_most).most_common(30)

# Convert to a DataFrame
common_words_df = pd.DataFrame(most_common, columns=['word', 'count'])

# Create a barplot
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='word', data=common_words_df, palette='viridis')

# Add a title
plt.title('Top 30 Most Common Words in Spam Messages', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Word', fontsize=12)
plt.show()


```

```python
ham_most = []

# Iterate through the list of spam messages (target == 1)
for msg in df[df['target'] == 0]['transform_text'].tolist():
    for word in msg.split():
        ham_most.append(word)

# Find the 30 most common words
most_common = Counter(ham_most).most_common(30)

# Convert to a DataFrame
common_words_df = pd.DataFrame(most_common, columns=['word', 'count'])

# Create a barplot
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='word', data=common_words_df, palette='viridis')

# Add a title
plt.title('Top 30 Most Common Words in Ham Messages', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Word', fontsize=12)
plt.show()

```

```python
## 4. Model Building
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv=CountVectorizer()
idf=TfidfVectorizer(max_features=3000)
```

```python
x=cv.fit_transform(df['transform_text'])
x=x.toarray()
x1=idf.fit_transform(df['transform_text'])
x1=x1.toarray()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y,test_size=0.2,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.metrics import accuracy_score,precision_score
```

```python
# defining all the objects for diffrent naive bayes models

gnb=GaussianNB()
bnb=BernoulliNB()
mnb=MultinomialNB()
```

```python
#gaussain nb for countvector
gnb.fit(x_train,y_train)

y_pred=gnb.predict(x_test)
print(accuracy_score(y_pred,y_test))
print(precision_score(y_pred,y_test))
```

```python
#gaussain nbfor tfidf
gnb.fit(x1_train,y1_train)
y1_pred=gnb.predict(x1_test)
print(accuracy_score(y1_pred,y1_test))
print(precision_score(y1_pred,y1_test))
```

```python
#multinomial nb for countvector
mnb.fit(x_train,y_train)

y_pred=mnb.predict(x_test)
print(accuracy_score(y_pred,y_test))
print(precision_score(y_pred,y_test))
```

```python
# multinomial nb for tfidfvector
mnb.fit(x1_train,y1_train)
y1_pred=mnb.predict(x1_test)
print(accuracy_score(y1_test,y1_pred))
print(precision_score(y1_pred,y1_test))
```

```python
bnb.fit(x_train,y_train)

y_pred=bnb.predict(x_test)
print(accuracy_score(y_pred,y_test))
print(precision_score(y_pred,y_test))
```

```python
bnb.fit(x1_train,y1_train)
y1_pred=bnb.predict(x1_test)
print(accuracy_score(y1_test,y1_pred))
print(precision_score(y1_pred,y1_test))
```
